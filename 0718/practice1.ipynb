```python
scores_df['總分'] = scores_df[['語文','英語', '數學', '地理', '歷史']].sum(axis = 1)
scores_df['平均'] = scores_df[['語文','英語', '數學', '地理', '歷史']].mean(axis = 1)
scores_df
```

```python
scores_df['排名'] = scores_df[['總分']].rank(ascending = False, method = 'min')
scores_df.sort_values(by = '排名')
```

```python
from pandas import Series:
def caculate (s:Series)-> int:
    n = (s < 60).sum()
    return n

scores_df['及格科目數'] = scores_df[['語文','英語', '數學', '地理', '歷史']].apply(caculate, axis =1)
```
