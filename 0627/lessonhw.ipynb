```python
# 請幫我用 Streamlit 製作一個圓餅圖（pie chart）的網頁應用程式，需求如下：

# 1. 使用 matplotlib 繪製圓餅圖
# 2. 資料內容：四個手機品牌 Nokia、Samsung、Apple、Lumia，對應數值分別為 20、30、45、10
# 3. 圓餅圖的顏色依序為黃色、綠色、紅色、藍色
# 4. 將第一塊（Nokia）使用 explode 突出顯示，突出距離 0.3
# 5. 加上陰影效果（shadow）
# 6. 每一塊要顯示百分比，格式為小數點後一位（例如 22.2%）
# 7. 起始角度設定為 180 度
# 8. 在網頁上方加一個標題文字「手機品牌市占率圓餅圖」
# 9. 使用 `st.pyplot()` 把圖表顯示在 Streamlit 網頁上


# 請給我完整的程式碼，並在最後告訴我如何用指令執行這個 Streamlit 應用程式。
```

```python
import streamlit as st
import matplotlib.pyplot as plt

st.title("手機品牌市占率圓餅圖")

labels = ["Nokia", "Samsung", "Apple", "Lumia"]
sizes = [20, 30, 45, 10]
colors = ["yellow", "green", "red", "blue"]
explode = [0.3, 0, 0, 0]

fig, ax = plt.subplots()

ax.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True,
    autopct="%.1f%%",
    startangle=180
)

ax.axis("equal")

st.pyplot(fig)
```
