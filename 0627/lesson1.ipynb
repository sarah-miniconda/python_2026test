```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi) 

figure = plt.figure()
ax1 = figure.add_subplot(2, 1, 1)
ax1.plot(t, y1, linestyle="--", color="red")
ax2 = figure.add_subplot(2, 1, 2)
ax2.plot(t, y2, linestyle="-.",color="blue")
plt.show()
```

```python
# 步驟 1:匯入 numpy,並取別名 np
import numpy as np
# 步驟 2:匯入 matplotlib.pyplot,並取別名 plt
import matplotlib.pyplot as plt
x = np.arange(0, 5 , 0.2)
# 步驟 4:計算 y1 = cos(pi * x)(用 np.cos 與 np.pi)
# 步驟 5:計算 y2 = cos(2 * pi * x)
y1 = np.cos(x * np.pi)
y2 = np.cos(2 * np.pi * x)

# 步驟 6:用 subplot 把版面切成 1 列 2 欄,選第 1 格(左圖)
# [X] ax1 = figure.add_subplot(2, 1, 1)
plt.subplot(1, 2, 1)
# 步驟 7:畫出 x 與 y1,使用綠色實線
# [X] ax1.plot(x, y1, linestyle="--", color="green")
plt.plot(x, y1, "g-")

# 步驟 8:用 subplot 選第 2 格(右圖)
plt.subplot(1, 2, 2)

# 步驟 9:畫出 x 與 y2,使用洋紅色點線
# [X]ax2.plot(x, y2, linestyle="--", color="red")
plt.plot(x, y2, "m:")

# 步驟 10:顯示圖形
plt.show()
```
