```python
# 學生總分300 有的可以加分5%

scores = int(input ("請輸入學生分數(最高分300分):"))
is_add = input("學生是否符合加分條件?(y, n)")

if is_add == "y": #單項選擇
    scores = int(scores + 0.5)
print("學生分數:",scores)
```
