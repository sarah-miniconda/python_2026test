```python
# 學生總分為300
#有些學生可以加分5%

while True:
# while ture 無限迴圈
# 縮排下面就是他的程式區塊
# 但是停不下來
# break >>跳出迴圈,例子在下面
# Ture的話下面continue >>回到上面的迴圈，繼續執行

    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        # raise value error，如果出現無法解決的問題
        break
    except ValueError:
        print("輸入錯誤，請輸入整數！")

is_add = input("學生是否符合加分條件?(y,n)")

if is_add == "y": #單項選擇
	scores = scores * 1.05

# 四捨五入(5進位)
scores = int(scores + 0.5)

# 超過300分，上限取300
if scores > 300:
	scores = 300

print("學生分數:",scores)
```

```python
# continue
i = 1

while i <= 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
# print出不能整除2的數字
```

```python
# continue & break
i = 0

while i < 10:
    i += 1
    if i == 3:
        continue
    if i == 8:
        break
    print(i)
```
