```python
# list
scores = [89, 92, 76, 91, 77]
# 運算式
print(scores)
print(scores[0])
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])

scores[0] = 100
print(scores)
```

```python
# tuple , 建立後不可改 ,暫時儲存
scores1 = (89, 92, 76, 91, 77)
print(scores1)

print(scores1[0])
print(scores1[1])
print(scores1[2])
print(scores1[3])
print(scores1[4])
```

```python
# dictionary
students = {'chinese':89, 'english':94, 'math':95, 'history':92, 'discover':95}
students = {
    'chinese':89, 
    'english':94, 
    'math':95, 
    'history':92, 
    'discover':95}
print(students)
print(students['chinese'])
print(students['english'])
print(students['math'])
print(students['history'])
print(students['discover'])
```
