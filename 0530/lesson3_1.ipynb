```python
# 引數值的呼叫，必須依照順序
try: 
    with open("student.txt", "r" , encoding = "utf-8")as file:
        data = file.read()
        print(data)
except Exception as e:
    print("發生錯誤:" , e)
```

```python
# open () 呼叫function
# open(file, mode(預設參數，可以省略)='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
# 'r' open for reading (default)
# 'w' open for writing, truncating the file first
# 'x' open for exclusive creation, failing if the file already exists
# 'a' open for writing, appending to the end of file if it exists
# 'b' binary mode
# 't' text mode (default)
# '+' open for updating (reading and writing)

```

```python
# 引數名稱的呼叫，可以不按照順序.
# file = "student.txt"

try: 
    with open (file = "student.txt", "r" , encoding = "utf-8")as file:
        data = file.read()
        print(data)
except Exception as e:
    print("發生錯誤:" , e)
```

```python
try: 
    with open (file = "student.txt") as file:
        data = file.read()
        print(data)
except Exception as e:
    print("發生錯誤:" , e)
```

```python
# 混合名稱的呼叫，前用引數值，後面引數名稱

try: 
    with open ("student.txt", encoding = "utf-8", mode = "r") as file:
        data = file.read()
        print(data)
except Exception as e:
    print("發生錯誤:" , e)
```
