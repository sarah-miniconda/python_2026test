```python
import csv

with open("考試分數_3年6班.csv", "r", encoding = "utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
        print(row["學生姓名"])# 抓每一個學生名字
    if int(row["數學"]) > 90:
        print(row["學生姓名"])
```
